```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Lambda, MaxPooling2D # convolution layers
from tensorflow.keras.layers import Dense, Dropout, Flatten # core layers

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
!cp /content/drive/MyDrive/analysis/spectograms/data4image_generator/utils.py /usr/local/lib/python3.7/dist-packages/keras_preprocessing/image/utils.py 
```

```python
datagen = ImageDataGenerator()
# load and iterate training dataset
train_it = datagen.flow_from_directory('/content/drive/MyDrive/analysis/spectograms/data4image_generator/train/', class_mode='binary', batch_size=64, target_size=(129, 48), color_mode='spec')
# load and iterate validation dataset
val_it = datagen.flow_from_directory('/content/drive/MyDrive/analysis/spectograms/data4image_generator/validation/', class_mode='binary', batch_size=64, target_size=(129, 48), color_mode='spec')
# load and iterate test dataset
test_it = datagen.flow_from_directory('/content/drive/MyDrive/analysis/spectograms/data4image_generator/test/', class_mode='binary', batch_size=3, target_size=(129, 48), color_mode='spec')
```

```python
def build_model(lr=0.01, momentum=0.9):
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(129, 48,16)))
    model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.2))
    model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
    model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.2))
    model.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
    model.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dropout(0.2))
    model.add(Dense(2, activation='softmax'))

    opt = tf.keras.optimizers.SGD(
        learning_rate=lr, momentum=momentum, nesterov=True,
    )
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics='binary_accuracy')
    return model
```

```python
model = build_model()
```

```python
tf.config.list_physical_devices('GPU') 
```

```python
history = model.fit_generator(train_it, steps_per_epoch=int(6956/64), validation_data=val_it, validation_steps=int(429/64), epochs=10)
```

```python
loss = model.evaluate_generator(test_it, steps=3)
loss
```

```python
yhat = model.predict_generator(test_it, steps=3)
yhat
```
